```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split

stats = pd.read_csv("NBA per 100 stats_cleaned.csv")
stats = stats.drop("Team", axis=1)
stats.info()
```

## Code Explanation

The provided code performs the following tasks:

1. **Library Imports**: It imports the necessary libraries for data manipulation, visualization, and modeling, including pandas, matplotlib.pyplot, seaborn, numpy, scikit-learn metrics, train_test_split, LinearRegression, and SimpleImputer.

2. **Data Loading**: It reads the data from the "NBA per 100 stats_cleaned.csv" file into a DataFrame called 'stats' using `pd.read_csv()`.

3. **Data Preparation**: It drops the "Team" column from the 'stats' DataFrame using `stats.drop("Team", axis=1)` to remove the unnecessary column.

4. **Data Overview**: It displays summary information about the 'stats' DataFrame using `stats.info()`. This provides an overview of the dataset, including the number of non-null values and data types for each column.

By utilizing these libraries and executing these steps, the code loads the dataset, prepares the data by removing the "Team" column, and offers a summary of the dataset's structure.



```python
X_train, X_test, y_train, y_test = train_test_split(
    stats.drop(columns=["PTS"]),
    stats["PTS"],
    test_size=0.2,
    random_state=42,
)
```

## Code Definition


The following tasks are carried out by the provided code:

1. Data Splitting: Using scikit-learn's 'train_test_split()' function, the dataset is divided into training and testing sets. The dependent variable (goal) is recorded in "y," whereas the independent variables (features) are kept in "X."

2. **Features and Target Variables**: It assigns training and testing sets for the features ('X_train' and 'X_test') as well as the target variable ('y_train' and 'y_test').

3. Data Splitting Parameters: The parameters for the data splitting operation are specified here:
   - "stats.drop(columns=["PTS"])" To generate the feature matrix, removes the "PTS" column from the "stats" DataFrame.
   -'stats["PTS"]': Chooses the "PTS" column from the'stats' DataFrame as the target variable.
   - 'test_size=0.2' explains that 80% of the data will be used for analysis and 20% will be used for testing.

```python
imputer = SimpleImputer(strategy="mean")
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```

## Code Definition

The following tasks are carried out by the provided code:


1. **Imputation**: To handle missing values in the data, it makes use of the 'SimpleImputer' class from scikit-learn.
2. Data transformation training: 'fit_transform()' is used to fit the imputer to the training data ('X_train'). This determines the mean value for each column in 'X_train' and substitutes the calculated means for any missing values.
3. **Transformation of Testing Data**: It uses the previously fitted imputer to transform the test data ('X_test') by executing 'transform()'. This substitutes the means computed from the data from the 'X_train' for any missing values in the 'X_test'.

The mean value of each individual column is used to replace missing values in the data when the "SimpleImputer" with the "mean" method is used. By doing this, the data is certain to be complete and prepared for analysis.

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

## Code Explanation

The provided code performs the following tasks:

1. **Model Initialization**: It creates an instance of the `LinearRegression` class from scikit-learn and assigns it to the variable `model`. This class represents a linear regression model.

2. **Model Training**: It fits the linear regression model (`model`) on the training data (`X_train` and `y_train`) using the `fit()` method. This process involves learning the coefficients and intercept of the linear regression equation that best fits the training data.

By creating a `LinearRegression` model and calling the `fit()` method, the code trains the model to learn the relationships between the independent variables (`X_train`) and the target variable (`y_train`).


```python
y_pred = model.predict(X_test)
```

## Code Explanation

The provided code performs the following tasks:

1. **Prediction**: It uses the trained linear regression model (`model`) to make predictions on the test data (`X_test`). The `predict()` method is called on the model with `X_test` as the input, and the predicted values are assigned to the variable `y_pred`.

By calling the `predict()` method on the trained linear regression model, the code generates predictions for the target variable (`y_pred`) based on the provided test data (`X_test`).


```python
score = model.score(X_test, y_test)
print(f"R^2 Score: {score}")
```

## Code Explanation

The provided code performs the following tasks:

1. **Model Evaluation**: It evaluates the performance of the linear regression model (`model`) on the test data (`X_test` and `y_test`) using the `score()` method. This method calculates the coefficient of determination (R-squared score) of the model.

2. **R-squared Score**: It assigns the calculated R-squared score to the variable `score`.

3. **Printing the Score**: It prints the R-squared score using the `print()` function and string formatting (`f"R^2 Score: {score}"`).

By using the `score()` method on the linear regression model, the code computes the R-squared score, which indicates how well the model fits the test data.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Code Explanation

The provided code performs the following tasks:

1. **Feature Selection**: It selects specific columns from the 'df_cleaned' DataFrame to form the feature matrix 'X'. The selected columns include 'MP' (minutes played), 'FG' (field goals made), 'FGA' (field goal attempts), '3P' (three-point field goals made), '3PA' (three-point field goal attempts), '2P' (two-point field goals made), '2PA' (two-point field goal attempts), 'FT' (free throws made), 'FTA' (free throw attempts), 'ORB' (offensive rebounds), 'DRB' (defensive rebounds), 'AST' (assists), 'STL' (steals), 'BLK' (blocks), 'TOV' (turnovers), and 'PF' (personal fouls).

2. **Target Variable Selection**: It selects the 'PTS' (points) column from the 'df_cleaned' DataFrame as the target variable 'y'.

3. **Data Splitting**: It splits the feature matrix 'X' and the target variable 'y' into training and testing sets using the `train_test_split()` function from scikit-learn. The training and testing sets are assigned to 'X_train', 'X_test', 'y_train', and 'y_test' respectively. The test size is set to 0.3 (30% of the data) and the random state is set to 101 for reproducibility.

By selecting specific columns as features and splitting the data into training and testing sets, the code prepares the dataset for model training and evaluation. The selected features are used to predict the target variable of points (PTS).

## Evaluating the Model

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

## Code Explanation

The provided code performs the following tasks:

1. **Mean Absolute Error (MAE)**: It calculates the mean absolute error between the true points (`y_test`) and the predicted points (`predictions`) using the `mean_absolute_error()` function from scikit-learn's metrics module. The MAE represents the average absolute difference between the actual and predicted values.

2. **Mean Squared Error (MSE)**: It calculates the mean squared error between the true points and the predicted points using the `mean_squared_error()` function from scikit-learn's metrics module. The MSE represents the average squared difference between the actual and predicted values.

3. **Root Mean Squared Error (RMSE)**: It calculates the root mean squared error by taking the square root of the MSE. The RMSE is a commonly used metric to assess the performance of a regression model, and it represents the square root of the average squared difference between the actual and predicted values.

By calculating the MAE, MSE, and RMSE, the code provides quantitative measures of the prediction accuracy of the linear regression model. These metrics help evaluate how well the model performs in predicting the points based on the selected features.
