```python
matches_df = pd.DataFrame(combined_df)
teams_df = pd.DataFrame(euro_melt)

# Initialize points columns for each prediction model
points_columns = ["points_xgboost", "points_rf", "points_qda", "points_average", "points_random"]
for column in points_columns:
    teams_df[column] = 0

# Function to update points
def update_points(df, home_team, away_team, winner, points_column):
    if winner == "Home":
        df.loc[df["Team"] == home_team, points_column] += 3
    elif winner == "Away":
        df.loc[df["Team"] == away_team, points_column] += 3
    elif winner == "Draw":
        df.loc[df["Team"] == home_team, points_column] += 1
        df.loc[df["Team"] == away_team, points_column] += 1

# Iterate over matches and update points
for index, match in matches_df.iterrows():
    for model in ["xgboost", "rf", "qda", "average", "random"]:
        update_points(teams_df, match["home_team"], match["away_team"], match[f"winner_{model}"], f"points_{model}")

# Function to get top 3 teams by predicted points
def get_top_three_teams_by_model(df, points_column):
    top_teams = df.groupby('Group').apply(lambda x: x.nlargest(3, points_column)).reset_index(drop=True)
    return top_teams[["Group", "Team", points_column]]

# Get top 3 teams for each prediction model
top_three_xgboost = get_top_three_teams_by_model(teams_df, "points_xgboost")
top_three_rf = get_top_three_teams_by_model(teams_df, "points_rf")
top_three_qda = get_top_three_teams_by_model(teams_df, "points_qda")
top_three_average = get_top_three_teams_by_model(teams_df, "points_average")
top_three_random = get_top_three_teams_by_model(teams_df, "points_random")

# Add model column to each dataframe
top_three_xgboost['model'] = 'xgboost'
top_three_rf['model'] = 'rf'
top_three_qda['model'] = 'qda'
top_three_average['model'] = 'average'
top_three_random['model'] = 'random'

# Concatenate all dataframes into one
combined_top_three = pd.concat([top_three_xgboost, top_three_rf[["Team",	"points_rf","model"]], top_three_qda[["Team",	"points_qda",	"model"]], top_three_average[["Team",	"points_average",	"model"]], top_three_random[["Team",	"points_random",	"model"]]],axis=1)

combined_top_three.reset_index(drop=True, inplace=True)
combined_top_three
```

```python
def determine_matchups(top_teams_list):
    matchups = []
    
    for model_df in top_teams_list:
        model = model_df['model'].iloc[0]
        
        # Get top teams in each group
        group_a = model_df[model_df['Group'] == 'Group_A'].sort_values(by=model_df.columns[2], ascending=False)
        group_b = model_df[model_df['Group'] == 'Group_B'].sort_values(by=model_df.columns[2], ascending=False)
        group_c = model_df[model_df['Group'] == 'Group_C'].sort_values(by=model_df.columns[2], ascending=False)
        group_d = model_df[model_df['Group'] == 'Group_D'].sort_values(by=model_df.columns[2], ascending=False)
        group_e = model_df[model_df['Group'] == 'Group_E'].sort_values(by=model_df.columns[2], ascending=False)
        group_f = model_df[model_df['Group'] == 'Group_F'].sort_values(by=model_df.columns[2], ascending=False)

        # Get top 3 teams in each group
        top_a = group_a.iloc[:3]['Team'].tolist()
        top_b = group_b.iloc[:3]['Team'].tolist()
        top_c = group_c.iloc[:3]['Team'].tolist()
        top_d = group_d.iloc[:3]['Team'].tolist()
        top_e = group_e.iloc[:3]['Team'].tolist()
        top_f = group_f.iloc[:3]['Team'].tolist()

        # Define matchups
        matchups.extend([
            (top_a[1], top_b[1], model),  # Second place Group A vs. Second place Group B
            (top_a[0], top_c[1], model),  # First place Group A vs. Second place Group C
            (top_c[0], top_d[2] if len(top_d) > 2 else top_e[2] if len(top_e) > 2 else top_f[2], model),  # First place Group C vs. 3rd in Group D, E or F
            (top_b[0], top_a[2] if len(top_a) > 2 else top_d[2] if len(top_d) > 2 else top_e[2] if len(top_e) > 2 else top_f[2], model),  # First place Group B vs. 3rd in Group A, D, E or F
            (top_d[1], top_e[1], model),  # Second place Group D vs. Second place Group E
            (top_f[0], top_a[2] if len(top_a) > 2 else top_b[2] if len(top_b) > 2 else top_c[2], model),  # First place Group F vs. 3rd in Group A, B or C
            (top_e[0], top_a[2] if len(top_a) > 2 else top_b[2] if len(top_b) > 2 else top_c[2] if len(top_c) > 2 else top_d[2], model),  # First place Group E vs. 3rd in Group A, B, C or D
            (top_d[0], top_f[1], model)   # First place Group D vs. Second place Group F
        ])

    return matchups

# Get matchups for the next round
next_round_matchups = determine_matchups([top_three_xgboost, top_three_rf,top_three_qda,top_three_average,top_three_random ])

# Display the combined dataframe and next round matchups
combined_top_three, next_round_matchups
```
